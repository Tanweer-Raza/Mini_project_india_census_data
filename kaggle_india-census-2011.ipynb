```python
final_df = latlong_df.merge(census_df,left_on ='District',right_on='District name').drop(columns=['District name'])
```

```python
final_df.head(2)
```

```python
final_df['literacy_rate'] = round((final_df['Literate']/final_df['Population'])*100)
final_df['Male_literacy_rate'] = round((final_df['Male_Literate']/final_df['Male'])*100)
final_df['Female_literacy_rate'] = round((final_df['Female_Literate']/final_df['Female'])*100)

final_df['sex_ratio'] = round((final_df['Female']/final_df['Male'])*1000)
```
